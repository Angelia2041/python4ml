```python
def check_pv(index_list):
    """
    Input a list of columns we are tring to summarize
    """
    pvsum=pd.pivot_table(df,index=index_list,aggfunc=np.sum)[["SHIPPED_QUANTITY","Sales_Rev","Sales_Cost"]]
    pvsum["Margin"]=pvsum["Sales_Rev"]-pvsum["Sales_Cost"]
    pvsum["Margin_Rate"]=pvsum["Margin"]/pvsum["Sales_Rev"]
    pvsum["Contrib_2_Margin"]=pvsum["Margin"]/pvsum["Margin"].sum()
    pvsum["Contrib_2_Sales"]=pvsum["Sales_Rev"]/pvsum["Sales_Rev"].sum()
    return pvsum
```

```python
check_pv(["SUPER_DEPT"]).sort_values(by="Contrib_2_Sales",ascending=False)
```

```python
check_pv(["SUPER_DEPT","DEPT"])
```

```python
from datetime import datetime
```

```python
wkdays=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
idx2wkd=dict((k,v) for k,v in enumerate(wkdays))
```

```python
df["WK_DT"]=list(df["EVENT_DT"].apply(lambda x:idx2wkd[datetime.strptime(str(x)[:10],"%Y-%m-%d").date().weekday()]))
```

### Clustering the sku

This is based on the phylosophy that if the sub categories sales goes up and down in 15 days in the similiar pattern, they can be considered 1 specific tpye of client, hence we can arrange the sub_category products under 1 cluster more closely on shelves or in closer places on website page.

```python
pvbyDay=pd.pivot_table(df[["SUB_CATEGORY","Sales_Rev","EVENT_DT"]],index=["SUB_CATEGORY"],aggfunc=np.sum,columns="EVENT_DT")
pvbyDay
```

```python
day_array=np.array(pvbyDay.fillna(0))
```

```python
# We normalize each row's sales data. Let price affect less of the buying pattern
rowtotal=day_array.sum(axis=1)
day_array_nor=day_array / rowtotal[:, np.newaxis]
```

```python
day_array_nor.shape
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=8, random_state=0)
```

```python
idx=kmeans.fit_predict(day_array_nor)
```

```python
pvbyDay["cluster_idx"]=list(idx)
```

```python
pvbyDay["cluster_idx"].value_counts()
```

### Using simple unsupervised learning to cluster sku together by their buying pattern

Kmeans is a good start to detect pattern. We cluster together the sub_category that with similiar buying pattern distributed in 15 days. Without specific user to item data, this is actually a lot to deduce the user habbits based only on sku details.

And it's subject to fluctuation caused by discounts.

```python
# The first big category
for i in list(pvbyDay[pvbyDay["cluster_idx"]==0].index):
    print(i,end="\t")
```
